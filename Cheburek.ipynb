```python
# Только числовые колонки
numeric_cols = df.select_dtypes(include=[np.number]).columns
# Если я знаю что все колонки числовые, то numeric_cols = df.columns  или вообще сразу в цикл вписать
# Словарь, чтобы сохранить результаты
outliers_info = {}

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Находим выбросы
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    num_outliers = len(outliers)

    # Сохраняем
    outliers_info[col] = {
        'count': num_outliers,
        'lower': lower_bound,
        'upper': upper_bound
    }

    if num_outliers > 0:
        print(f"В колонке '{col}' найдено {num_outliers} выбросов")

# Можно посмотреть всё в виде таблицы
outliers_df = pd.DataFrame(outliers_info).T
print(outliers_df)
```

```python
# Размер группы (сколько колонок выводить за раз)
group_size = 10

# Разбиваем на группы
groups = [numeric_cols[i:i + group_size] for i in range(0, len(numeric_cols), group_size)]
print(f"Количество групп: {len(groups)}")

# Перебираем группы
for i, group in enumerate(groups):
    # Создаём фигуру (сетку графиков)
    fig, axes = plt.subplots(
        nrows=2, ncols=5,  # 2 строки, 5 столбцов = 10 графиков
        figsize=(15, 6)    # размер всей фигуры
    )

    # "Плоский" список осей (чтобы проще перебирать)
    axes = axes.flatten()

    # Перебираем колонки в группе
    for idx, col in enumerate(group):
        # Строим гистограмму для текущей колонки
        df[col].hist(ax=axes[idx], bins=30, alpha=0.7)
        axes[idx].set_title(col)  # заголовок — название колонки

    # Убираем пустые оси (если в группе меньше 10 колонок)
    for j in range(len(group), len(axes)):
        fig.delaxes(axes[j])

    # Улучшаем отступы
    plt.tight_layout()
    plt.suptitle(f'Группа {i+1}', y=1.02)  # заголовок группы
    plt.show()
```

```python
# Функция для расчёта IV по одной фиче
def calculate_iv(df, feature, target, bins=10):
    df = df[[feature, target]].copy()

    # Если фича числовая — бинируем
    try:
        df['bin'] = pd.qcut(df[feature], q=bins, duplicates='drop')
    except:
        df['bin'] = df[feature]  # если qcut не сработал — категориальная фича

    # Считаем good/bad
    grouped = df.groupby('bin', observed=False)[target].agg(['count', 'sum'])
    grouped.columns = ['total', 'bad']
    grouped['good'] = grouped['total'] - grouped['bad']

    # Распределения good/bad
    dist_good = grouped['good'] / grouped['good'].sum()
    dist_bad = grouped['bad'] / grouped['bad'].sum()

    # WoE и IV
    grouped['woe'] = np.log((dist_good + 1e-9) / (dist_bad + 1e-9))  # +1e-9 чтобы избежать деления на 0
    grouped['iv'] = (dist_good - dist_bad) * grouped['woe']

    return grouped['iv'].sum()

# Считаем IV для всех признаков
iv_scores = {}

for col in df.columns:
    if col != 'target':
        try:
            iv_scores[col] = calculate_iv(df, col, 'target')
        except Exception as e:
            print(f"{col}: ошибка — {e}")

# Сортируем по убыванию
iv_sorted = pd.Series(iv_scores).sort_values(ascending=False)

# Покажем топ-20
print("Top 20 признаков по IV:")
print(iv_sorted.head(20))

# Визуализация
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x=iv_sorted.head(20).values, y=iv_sorted.head(20).index, palette='viridis')
plt.title("Top 20 признаков по Information Value (IV)")
plt.xlabel("IV")
plt.ylabel("Признаки")
plt.tight_layout()
plt.show()
```

```python
# Импорты
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Подготовка данных - ИСПРАВЛЕНО
print("Подготовка данных...")
# Отделяем признаки и таргет
X = df.drop('target', axis=1)
y = df['target']

# Проверка на пропуски
print(f"Пропусков в признаках: {X.isnull().sum().sum()}")
print(f"Пропусков в таргете: {y.isnull().sum()}")

# Заполнение пропусков (если остались)
if X.isnull().sum().sum() > 0:
    X = X.fillna(X.median())  # Используем медиану

# ИСПРАВЛЕНИЕ: Преобразование интервальных данных в числовые
print("Проверка на интервальные данные...")
for col in X.columns:
    if hasattr(X[col].iloc[0] if len(X[col]) > 0 else None, 'left'):
        # print(f"Найден интервальный столбец: {col}")
        # Преобразуем интервал в его середину
        X[col] = X[col].apply(lambda x: (x.left + x.right) / 2 if pd.notna(x) else np.nan)

# Проверка типов данных
print("Типы данных в признаках:")
print(X.dtypes.value_counts())

# Преобразование категориальных данных в числовые (если есть)
categorical_cols = X.select_dtypes(include=['object', 'category']).columns
if len(categorical_cols) > 0:
    print(f"Найдены категориальные столбцы: {list(categorical_cols)}")
    # Можно использовать pd.get_dummies() или LabelEncoder
    X = pd.get_dummies(X, columns=categorical_cols, dummy_na=True)

# Проверка формы данных
print(f"Форма X: {X.shape}")
print(f"Форма y: {y.shape}")

# Стандартизация признаков
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# Разделение на train/test
print("Разделение данных...")
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)


print(f"Размер обучающей выборки: {X_train.shape}")
print(f"Размер тестовой выборки: {X_test.shape}")

# Проверка баланса классов
print(f"Распределение классов в train: {np.bincount(y_train)}")
print(f"Распределение классов в test: {np.bincount(y_test)}")

# Обучение логистической регрессии с L1-регуляризацией
print("Обучение модели...")
try:
    # Создаем модель
    log_reg = LogisticRegression(penalty='l1', solver='liblinear', random_state=42, max_iter=1000)

    # Подбор оптимального C
    param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}
    grid_search = GridSearchCV(log_reg, param_grid, cv=5, scoring='roc_auc')
    grid_search.fit(X_train, y_train)

    # Лучшая модель
    best_log_reg = grid_search.best_estimator_
    print(f"Оптимальное значение C: {grid_search.best_params_['C']}")

except Exception as e:
    print(f"Ошибка при обучении: {e}")
    # Попробуем с другими параметрами
    best_log_reg = LogisticRegression(penalty='l1', solver='liblinear', C=1.0, random_state=42, max_iter=1000)
    best_log_reg.fit(X_train, y_train)

# Важность признаков
print("Анализ признаков...")
try:
    coef_series = pd.Series(best_log_reg.coef_[0], index=X.columns)
    selected_features = coef_series[coef_series != 0]

    print(f"Всего признаков: {len(coef_series)}")
    print(f"Отобранных признаков: {len(selected_features)}")
    print("\nТоп-10 признаков по важности:")
    print(selected_features.sort_values(key=abs, ascending=False).head(10))

except Exception as e:
    print(f"Ошибка при анализе коэффициентов: {e}")

# Предсказание и оценка
print("Оценка модели...")
try:
    # Предсказания
    y_pred_proba = best_log_reg.predict_proba(X_test)[:, 1]
    y_pred = best_log_reg.predict(X_test)

    # Метрики
    accuracy = accuracy_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    print(f"Accuracy: {accuracy:.4f}")
    print(f"ROC-AUC: {roc_auc:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

except Exception as e:
    print(f"Ошибка при оценке: {e}")

# Визуализации
try:
    # ROC кривая
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.3f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()

    # Confusion Matrix
    plt.subplot(1, 2, 2)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')

    plt.tight_layout()
    plt.show()

    # Важность признаков
    if len(selected_features) > 0:
        plt.figure(figsize=(10, max(6, len(selected_features) * 0.3)))
        selected_features.sort_values(key=abs).plot(kind='barh')
        plt.title("Коэффициенты логистической регрессии с L1 (Lasso)")
        plt.xlabel("Вес признака")
        plt.tight_layout()
        plt.show()

except Exception as e:
    print(f"Ошибка при визуализации: {e}")

print("Анализ завершен!")
```
